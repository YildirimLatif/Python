```python
#liste üreteçleri: for döngüsü
x = []
for i in range(0,25):
    if i%2==0:
        x.append(i)
    else:
        x.append(i*i)
print(x)
```

```python
#Tek satıra indirgeme:
x = [i if i%2==0 else i*i for i in range(0,25)]
print(x)
```

```python
#liste üreteçleri: for döngüsü
x = []
for i in range(0,25):
    if i%2==0:
        x.append(i)

print(x)
```

```python
#Tek satıra indirgeme:
x = [i for i in range(0,25) if i%2==0 ]
print(x)
```

```python
def mod_func(n):
    return lambda x : x**n

pow_2 = mod_func(2)
pow_3 = mod_func(3)
pow_4 = mod_func(4)

print(type(pow_2))

print(pow_2(2))
print(pow_3(2))
print(pow_4(2))
```

```python
def mod_func(n):
    return lambda x : x**n

pow_2 = mod_func(2)(2)
pow_3 = mod_func(3)(2)
pow_4 = mod_func(4)(2)

print(pow_2)
print(pow_3)
print(pow_4)

print(type(pow_4))
```

```python
def mod_func(n):
    return lambda x : x*n

pow_2 = mod_func(2)("ali ")
pow_3 = mod_func(3)("alex ")
pow_4 = mod_func(4)("serpil ")

print(pow_2)
print(pow_3)
print(pow_4)

print(type(pow_4))
```

```python
def functioner(emoji=None):
    return lambda msj : print(msj, emoji)

_smile = functioner(":)")
_smile("hello")
_sad = functioner(":(")
_sad("hello")
```
