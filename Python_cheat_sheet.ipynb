### **STRING IFADE KOD KULLANIMLARI**

```python
konu = '  Havalar iyice, sogudu Murat beyyy  '
text = 'Muratterzi'
text += ' adam'                                     #   string ifade olarak eklemek icin kullanilir
print('1:', konu.swapcase())                        #   b√ºy√ºk ve k√ºy√ºk harfleri tam tersi yapar
print('2:', konu.upper())                           #   hepsini b√ºy√ºk yapar
print('3:', konu.lower())                           #   hepsini k√ºc√ºk yapar
print('4:', konu.casefold())                        #   hepsini k√ºc√ºk yapar
print('5:', konu.capitalize())                      #   Sadece satir basi b√ºy√ºk olur
print('6:', konu.title())                           #   T√ºm kelimeler b√ºy√ºk ile baslar
print('7:', konu.replace('r', 'g',2))               #   c√ºmle icinden istenilen harfi baska bir harf ile degistirir.(sayi kac kere degistirecegini belirler)
print('8:', konu.count('a', 5, 25))                 #   c√ºmlede istenilen harften kac tane var onu sayar veya kac karakter var onu sayar. (start ve stop indeksi verilebilir)
print('9:', konu.startswith('Hava', 2))             #   c√ºmle istenilen kelimeler ile basliyorsa " True veya False " verir. (start ve stop indeksi verilebilir)
print('10:', konu.endswith('bey', 0, 32))           #   c√ºmle istenilen kelimeler ile bitiyorsa " True veya False " verir. (start ve stop indeksi verilebilir)
print('11:', konu.startswith('i', 8))               #   c√ºmle icinden konum vererek harf buldurulur " True ve False " alinir
print('12:', konu.strip(' y '))                     #   c√ºmlenin iki tarafindaki bosluklari veya icindeki harflari kaldirir
print('13:', konu.lstrip(' Ha'))                    #   soldan istedigimiz kelimeleri siler (bosluklar √∂nemli) "kac tane o kelimeden bulursa siler"
print('14:', konu.rstrip('ey '))                    #   sagdan istedigimiz kelimeri siler (bosluklar √∂nemli) "kac tane o kelimeden bulursa siler"
print('15:', konu.find('a'))                        #   c√ºmle icinden istenilen harfin konumunu soldan arar ve ilk buldugu degeri verir
print('16:', konu.rfind('t'))                       #   c√ºmle icinden istenilen harfin konumunu sagdan arar ve ilk buldugu degeri verir
print('17:', text.index('a'))                       #   c√ºmle icinden istenilen harfin konumunu verir (b√ºy√ºk k√ºc√ºk harf duyarli). (0 dan baslar)
print('18:', *konu)                                 #   " * " harflerin arasina bosluk ekler
print('19:', 'iyi' in konu)                         #   icinde istenilen harf var mi diye bakar (True ve False verir)
print('20:', '@' not in konu)                       #   icinde istenilen harf yok mu diye bakar (True ve False verir)
print('21:', ' # '.join(konu))                      #   her karakter sonrasina istenilen karakteri ekler.
print('22:', konu.center(46, '+'))                  #   istenilen karakter sayisi dolana kadar sola ve saga ekleme yapar
print('23:', konu.ljust(40, '.'))                   #   yaziyi sola yaslar ve saga istenilen sayi kadar karakter ile doldurur.
print('24:', konu.rjust(50, '/'))                   #   yaziyi saga yaslar ve sola istenilen sayi kadar karakter ile doldurur.
print('25:', konu.zfill(45))                        #   yazinin basina istenilen sayi kadar "0" sifir ekler      
print('26:', konu.split('a'))                       #   ayristirma yapiyor, icine girilen karakteri kaldiriyor ve virg√ºl ile ayiriyor.
print('27:', konu.split(',')[0])                    #   icine girilen karakterden itibaren ayirdik ve indis vererek yazdirdik.
print('28:', min(konu), max(konu))                  #   ASCII kod tablosuna g√∂re en son ve ilk harfi yazdirir
print('29:', text.isalpha())                        #   tamami harflerden mi olusuyor diye sorgulama yaptik       
print('30:', text.isdigit())                        #   tamami rakamlardan mi olusuyor diye sorgulama yaptik
print('30:', text.islower())                        #   tamami k√ºc√ºk harf mi olusuyor diye sorgulama yaptik
print('31:', konu.isalnum())                        #   rakam ve harften mi olusuyor diye sorgulama yaptik. (√ñzel simgeleri kabul etmez)
print('32:', text)
```
